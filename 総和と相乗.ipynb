総和を実装

```python
import numpy as np

a = np.array([1, 3, 2, 5, 4]) #a1からa5まで
y = np.sum(a) #総和
print(y)
```

総乗を実装

```python
import numpy as np

a = np.array([1, 3, 2, 5, 4]) #a1からa5まで
y = np.prod(a) #総乗
print(y)
```

演習

```python
import numpy as np

b = np.array([6, 1, 5, 4, 3, 2])

print(np.sum(b)) #総和
print(np.prod(b)) #総乗
```
