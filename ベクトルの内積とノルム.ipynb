内積の実装

```python
import numpy as np

a = np.array([1, 2, 3])
b = np.array([3, 2, 1])

print("…dot()関数…")
print(np.dot(a, b)) #dot()関数による内積
print("…積の総和…")
print(np.sum(a*b)) #積の総和による内積
```

ノルムの実装

```python
import numpy as np

a = np.array([1, 1, -1, -1])

print("…1.2 ノルム…")
print(np.linalg.norm(a)) #1.2 ノルム（デフォルト）
print("…L1 ノルム…")
print(np.linalg.norm(a, 1)) #L1 ノルム
```

演習

```python
import numpy as np

a = np.array([1, -2, 2])
b = np.array([2, -2, 1])

print("…内積…")
print(np.dot(a, b))

print("…1.2 ノルム…")
print(np.linalg.norm(a))

print("…L1 ノルム…")
print(np.linalg.norm(a, 1))
```
