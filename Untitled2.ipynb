```python
# データの例
data = [10, 15, 20, 25,30]

# 平均を計算
mean = sum(data) / len(data)
print("平均:", mean)

# 平均偏差を計算
mean_deviation = sum(abs(x - mean) for x in data) / len(data)
print("平均偏差:", mean_deviation)
```

```python
import numpy as np

data = np.array([10, 15, 20, 25, 30])

mean = np.mean(data)
print("平均:", mean)

mean_deviation = np.mean(np.abs(data - mean))
print("平均偏差:", mean_deviation)
```

```python
# データ
data = [10, 15, 20, 25, 30]

# 平均
mean = sum(data) / len(data)

# 分散
variance = sum((x - mean) ** 2 for x in data) / len(data)

print("平均:", mean)
print("分散:", variance)
```

```python
import numpy as np

data = np.array([10, 15, 20, 25, 30])

variance = np.var(data)

print("分散:", variance)
```

```python
data_A = [10,12,15,20]
data_B = [8, 9, 11, 13, 14]

def mean_deviation(data):
    mean = sum(data) / len(data)
    abs_dev = [abs(x - mean) for x in data]
    return sum(abs_dev) / len(abs_dev)

mad_A = mean_deviation(data_A)
mad_B = mean_deviation(data_B)

mad_A, mad_B
```

```python
import matplotlib.pyplot as plt

名前 = ['A', 'B', 'C', 'D', 'E']
英語 = [90, 80, 20, 50, 60]
数学 = [70, 10, 80, 70, 60]

英語 = [float(x) for x in 英語]
数学 = [float(x) for x in 数学]

plt.scatter(英語, 数学, color='blue')

for i, n in enumerate(名前):
    plt.text(英語[i]+0.3, 数学[i]+0.3, n)

plt.title("英語と数学の点数　(名前付き)　")
plt.xlabel("英語の点数")
plt.ylabel("数学の点数")

plt.show()
```
