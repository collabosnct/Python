# 第3回：リスト、タプル、繰り返し文、文字列

複数のデータをまとめて扱う方法を学びます。

## 1. リストとタプル
複数のデータを順番に並べて保存するためのデータ構造です。要素の数え方（インデックス）は `0` から始まります。
* **リスト `[]`**: プログラムの途中で要素を追加、削除、書き換えができる柔軟な構造です。

* **タプル `()`**: 一度作成すると、要素の書き換えができない（変更不可）構造です。誤って書き換えたくない重要なデータを扱う際に使います。


## 2. 繰り返し文 (`for`, `while`)
* **`for`文**: リストやタプルなどからデータを1つずつ取り出し、すべての要素に対して処理を繰り返すときに使います。


* **`while`文**: 「ある条件が `True` である間」ずっと処理を繰り返すときに使います。



## 3. 文字列の操作
文字列も「文字が順番に並んだリストのようなもの」として扱うことができます。長さを調べたり、一部を切り出したり（スライス）することが可能です。

```python
# --- 1. リストとタプル ---
# リスト（変更可能）
scores_list = [70, 85, 90]
scores_list[0] = 75  # 0番目の要素を上書き
scores_list.append(100)  # 末尾に新しい要素を追加

# タプル（変更不可）
weekdays_tuple = ("月", "火", "水", "木", "金")
# weekdays_tuple[0] = "日" # タプルの内容を変更しようとするとエラーになります

print("--- データ構造 ---")
print("更新されたリスト:", scores_list)
print("タプルのデータ:", weekdays_tuple)

# --- 2. 繰り返し文 ---
print("\n--- for文による繰り返し ---")
total_score = 0
for score in scores_list:
    print(f"得点: {score}")
    total_score = total_score + score

print(f"合計点は {total_score} 点です。")

# --- 3. 文字列の操作 ---
print("\n--- 文字列（スライスなど） ---")
message = "PythonProgramming"

print("元の文字:", message)
print("文字の長さ:", len(message))
print("最初の6文字を切り出し:", message[0:6])  # 0番目から6番目の手前(5番目)まで
print("大文字に変換:", message.upper())
```
