ğŸ¯**ç›®çš„**

AIï¼ˆæ©Ÿæ¢°å­¦ç¿’ï¼‰ã®åŸºæœ¬ã§ã‚ã‚‹ã€Œãƒ‡ãƒ¼ã‚¿ã‹ã‚‰æ³•å‰‡ï¼ˆãƒ‘ãƒ©ãƒ¡ãƒ¼ã‚¿ï¼‰ã‚’è¦‹ã¤ã‘å‡ºã™ã€ãƒ—ãƒ­ã‚»ã‚¹ã‚’ä½“é¨“ã—ã¾ã™ã€‚ä»Šå›ã¯ã€`scikit-learn` ãªã©ã®ä¾¿åˆ©ãªãƒ©ã‚¤ãƒ–ãƒ©ãƒªã‚’ä½¿ã‚ãšã€**NumPy**ã ã‘ã‚’ä½¿ã£ã¦ã€ãƒ‡ãƒ¼ã‚¿ã®å‚¾å‘ã‚’è¡¨ã™ç›´ç·š y = ax + b ã‚’è¨ˆç®—ã§æ±‚ã‚ã¾ã™ã€‚

ğŸ–Š**å‰æçŸ¥è­˜**



*   **å˜å›å¸°ãƒ¢ãƒ‡ãƒ«**ï¼šy = ax + b
*   **ç›®çš„**ï¼šãƒ‡ãƒ¼ã‚¿ã¨ã®èª¤å·®ãŒæœ€å°ã«ãªã‚‹ aï¼ˆå‚¾ãï¼‰ã¨ bï¼ˆåˆ‡ç‰‡ï¼‰ã‚’è¦‹ã¤ã‘ã‚‹ã“ã¨ã€‚
*   **æœ€å°äºŒä¹—æ³•**ï¼šèª¤å·®ã®äºŒä¹—å’Œã‚’æœ€å°ã«ã™ã‚‹ãƒ‘ãƒ©ãƒ¡ãƒ¼ã‚¿ã‚’æ±‚ã‚ã‚‹æ•°å­¦çš„æ‰‹æ³•ã€‚


**ã€æº–å‚™ã€‘ãƒ‡ãƒ¼ã‚¿ã®ç”Ÿæˆ**

ã¾ãšã¯åˆ†æã™ã‚‹ãŸã‚ã®æ¶ç©ºã®ãƒ‡ãƒ¼ã‚¿ã‚’ä½œã‚Šã¾ã™ã€‚ä»¥ä¸‹ã®ã‚³ãƒ¼ãƒ‰ã‚’å®Ÿè¡Œã—ã¦ãã ã•ã„ã€‚ï¼ˆã“ã“ã§ã¯ã€Œå‹‰å¼·æ™‚é–“ã€ã¨ã€Œãƒ†ã‚¹ãƒˆã®ç‚¹æ•°ã€ã®é–¢ä¿‚ã‚’ã‚·ãƒŸãƒ¥ãƒ¬ãƒ¼ã‚·ãƒ§ãƒ³ã—ã¾ã™ï¼‰

```python
import numpy as np
import matplotlib.pyplot as plt

#ãƒ©ãƒ³ãƒ€ãƒ ã‚·ãƒ¼ãƒ‰ã‚’å›ºå®šï¼ˆå…¨å“¡åŒã˜ãƒ‡ãƒ¼ã‚¿ã«ãªã‚‹ã‚ˆã†ã«ï¼‰
np.random.seed(42)

#xï¼šå‹‰å¼·æ™‚é–“ï¼ˆ0æ™‚é–“ï½10æ™‚é–“ï¼‰ã‚’20äººåˆ†ç”Ÿæˆ
x = 10 * np.random.rand(20)

#yï¼šãƒ†ã‚¹ãƒˆã®ç‚¹æ•°ï¼ˆå‹‰å¼·æ™‚é–“Ã—8ï¼‹10ç‚¹ï¼‹èª¤å·®ãƒã‚¤ã‚ºï¼‰
#æ­£è§£ã®æ³•å‰‡ã¯ a=8ï¼Œb=10ã§ã™ãŒã€AIã«ã¯ã“ã‚Œã‚’ãƒ‡ãƒ¼ã‚¿ã‹ã‚‰å½“ã¦ã•ã›ã¾ã™
y = 8 * x + 10 + np.random.randn(20) * 5

print("ãƒ‡ãƒ¼ã‚¿ç”Ÿæˆå®Œäº†")
print(f"x(å‹‰å¼·æ™‚é–“):{x[:5]}...")
print(f"y(ç‚¹æ•°):{y[:5]}...")
```

**èª²é¡Œï¼‘ï¼šãƒ‡ãƒ¼ã‚¿ã®å¯è¦–åŒ–**

ãƒ‡ãƒ¼ã‚¿ãŒã©ã‚“ãªå½¢ã‚’ã—ã¦ã„ã‚‹ã‹ç¢ºèªã—ã¾ã—ã‚‡ã†ã€‚`matplotlib` ã‚’ä½¿ã£ã¦ã€æ¨ªè»¸ã« x ã€ç¸¦è»¸ã« y ã‚’ã¨ã£ãŸ**æ•£å¸ƒå›³**ã‚’æç”»ã—ã¦ãã ã•ã„ã€‚

```python
#ã€èª²é¡Œï¼‘ã€‘ã“ã“ã«ã‚³ãƒ¼ãƒ‰ã‚’è¨˜è¿°
#ãƒ’ãƒ³ãƒˆï¼šplt.scatterï¼ˆæ¨ªè»¸ï¼Œç¸¦è»¸ï¼‰ä½¿ã„ã¾ã™
plt.scatter(x, y)
plt.xlabel("Study Hours")
plt.ylabel("Score")
plt.grid(True)
plt.show()
```

**èª²é¡Œï¼’ï¼šäºˆæ¸¬é–¢æ•°ã®å®Ÿè£…**

ãƒ‘ãƒ©ãƒ¡ãƒ¼ã‚¿ aï¼ˆå‚¾ãï¼‰ã¨bï¼ˆåˆ‡ç‰‡ï¼‰ã‚’å—ã‘å–ã‚Šã€å…¥åŠ›ãƒ‡ãƒ¼ã‚¿ x ã«å¯¾ã™ã‚‹äºˆæ¸¬å€¤ `y_pred`ã‚’è¿”ã™é–¢æ•° `predict(x, a, b) `ã‚’ä½œæˆã—ã¦ãã ã•ã„ã€‚æ•°å¼ï¼š$\hat{y}$ = ax + b

```python
#ã€èª²é¡Œï¼’ã€‘é–¢æ•°ã‚’å®Œæˆã•ã›ã¦ãã ã•ã„
def predict(x, a, b):
    #ã“ã“ã«å®Ÿè£…
    return a * x + b
```

**èª²é¡Œï¼“ï¼šèª¤å·®é–¢æ•°ï¼ˆMSEï¼‰ã®å®Ÿè£…**

ã€Œäºˆæ¸¬ã€ã¨ã€Œæ­£è§£ã€ãŒã©ã‚Œãã‚‰ã„ã‚ºãƒ¬ã¦ã„ã‚‹ã‹ã‚’è¨ˆç®—ã™ã‚‹é–¢æ•° `mse(y_true, y_pred)` ã‚’ä½œæˆã—ã¦ãã ã•ã„ã€‚ä»Šå›ã¯**å¹³å‡äºŒä¹—èª¤å·® (Mean Squared Error)** ã‚’ä½¿ã„ã¾ã™ã€‚

æ•°å¼ï¼š$$
\text{MSE} = \frac{1}{N} \sum_{i=1}^{N} \left( y_{\text{true}}^{(i)} - y_{\text{pred}}^{(i)} \right)^2
$$

*ãƒ’ãƒ³ãƒˆï¼šNumpyã® `np.mean()` ã‚„ `** 2`ï¼ˆäºŒä¹—ï¼‰ã‚’æ´»ç”¨ã™ã‚‹ã¨1è¡Œã§æ›¸ã‘ã¾ã™ã€‚*

```python
#ã€èª²é¡Œï¼“ã€‘é–¢æ•°ã‚’å®Œæˆã•ãˆã¦ãã ã•ã„
def mse(y_true, y_pred):
    #ã“ã“ã«å®Ÿè£…
    #è¦ç´ ã”ã¨ã®å·®ã®äºŒä¹—ã®å¹³å‡
    return np.mean((y_true - y_pred) ** 2)
```

**ã€èª²é¡Œï¼”ã€‘æœ€å°äºŒä¹—æ³•ã®å…¬å¼ã§å­¦ç¿’ï¼ˆãƒ¡ã‚¤ãƒ³èª²é¡Œï¼‰**

ã„ã‚ˆã„ã‚ˆãƒ‡ãƒ¼ã‚¿ã‹ã‚‰æœ€é©ãª a ã¨ b ã‚’æ±‚ã‚ã¾ã™ã€‚ä»Šå›ã¯å‹¾é…é™ä¸‹æ³•ï¼ˆæ¬¡å›æ‰±ã„ã¾ã™ï¼‰ã§ã¯ãªãã€ **æ•°å­¦çš„ãªå…¬å¼ï¼ˆæ­£è¦æ–¹ç¨‹å¼ã®ç°¡æ˜“ç‰ˆï¼‰** ã‚’ãã®ã¾ã¾ã‚³ãƒ¼ãƒ‰ã«æ›¸ã„ã¦ä¸€ç™ºã§æ±‚ã‚ã¾ã™ã€‚

çµ±è¨ˆå­¦ã«ã‚ˆã‚Šã€å˜å›å¸°åˆ†æã®æœ€é©ãª a ã¨ b ã¯ä»¥ä¸‹ã®å¼ã§æ±‚ã¾ã‚‹ã“ã¨ãŒã‚ã‹ã£ã¦ã„ã¾ã™ã€‚

ã“ã“ã§ã€
*   $\bar{x}$, $\bar{y}$ã¯ãã‚Œãã‚Œã®å¹³å‡å€¤ (`np.mean`)
*   åˆ†æ•£ (x) ã¯ `np.var`(`x, ddof=0`)
*   å…±åˆ†æ•£ (x, y) ã¯ `np.cov`(`x, y, ddof=0`)[`0, 1`]

ã‚’ä½¿ã£ã¦è¨ˆç®—ã§ãã¾ã™ã€‚ã“ã®å…¬å¼ã‚’Pythonã‚³ãƒ¼ãƒ‰ã«å¤‰æ›ã—ã¦ãã ã•ã„ã€‚


```python
#ã€èª²é¡Œï¼”ã€‘æœ€é©ãª a ã¨ b ã‚’è¨ˆç®—ã—ã¦ãã ã•ã„

#1. xã¨yã®å¹³å‡ã‚’è¨ˆç®—
mu_x = np.mean(x)
mu_y = np.mean(y)

#2. xã®åˆ†æ•£ã¨ã€xã¨yã®å…±åˆ†æ•£ã‚’è¨ˆç®—
#â€»np.varã¯ãƒ‡ãƒ•ã‚©ãƒ«ãƒˆã§æ¨™æœ¬åˆ†æ•£ã§ã™ãŒã€ã“ã“ã§ã¯åˆ†æ¯ã®NãŒã‚­ãƒ£ãƒ³ã‚»ãƒ«ã•ã‚Œã‚‹ãŸã‚ã‚ã¾ã‚Šæ°—ã«ã—ãªãã¦ã‚ˆã„
var_x = np.var(x, ddof=0)
cov_xy = np.mean((x - mu_x) * (y - mu_y)) #å…±åˆ†æ•£ã‚’æ‰‹è¨ˆç®—ã™ã‚‹å ´åˆ
#ã¾ãŸã¯ cov_xy = np.cov(x, y, ddof=0)[0, 1]

#ãƒ‘ãƒ©ãƒ¡ãƒ¼ã‚¿è¨ˆç®—
#3. å…¬å¼é€šã‚Šã« a ã‚’æ±‚ã‚ã‚‹
a_opt = cov_xy / var_x

#4. å…¬å¼é€šã‚Šã« b ã‚’æ±‚ã‚ã‚‹
b_opt = mu_y - a_opt * mu_x

print(f"AIãŒå°ãå‡ºã—ãŸæ³•å‰‡ï¼šç‚¹æ•° = {a_opt:.2f} Ã— å‹‰å¼·æ™‚é–“ + {b_opt:.2f}")
#æ­£è§£(a=8, b=10)ã«è¿‘ã„å€¤ãŒå‡ºã¦ã„ã‚‹ã¯ãšã§ã™ï¼
```

**èª²é¡Œï¼•ï¼šçµæœã®ç¢ºèª**

æœ€å¾Œã«ã€æ±‚ã‚ãŸ a, b ã‚’ä½¿ã£ã¦ç›´ç·šã‚’å¼•ã„ã¦ã¿ã¾ã—ã‚‡ã†ã€‚æ•£å¸ƒå›³ã®ä¸Šã«ã€AIãŒå­¦ç¿’ã—ãŸå›å¸°ç›´ç·šã‚’é‡ã­ã¦è¡¨ç¤ºã—ã¦ãã ã•ã„ã€‚

```python
#ã€èª²é¡Œï¼•ã€‘æ•£å¸ƒå›³ã¨å›å¸°ç›´ç·šã‚’é‡ã­ã¦è¡¨ç¤º

#1. æ•£å¸ƒå›³ã‚’æç”»
plt.scatter(x, y, label='Data', color='blue')

#2. å›å¸°ç›´ç·šã‚’è¨ˆç®—ï¼ˆxã®æœ€å°å€¤ã‹ã‚‰æœ€å¤§å€¤ã¾ã§ã®ç¯„å›²ã§ç›´ç·šã‚’å¼•ããŸã‚ã®ãƒ‡ãƒ¼ã‚¿ã‚’ä½œã‚‹ï¼‰
x_line = np.linspace(0, 10, 100)
y_line = predict(x_line, a_opt, b_opt)

#3. ç›´ç·šã‚’æç”»ï¼ˆplt.plotã‚’ä½¿ã†ï¼‰
plt.plot(x_line, y_line, color='red', label='AI Prediction', linewidth=2)

plt.legend()
plt.xlabel("Study Hours")
plt.ylabel("Test Score")
plt.grid(True)
plt.show()
```
