固有値と固有ベクトルの計算

```python
import numpy as np

a = np. array([[3, 1], [2, 4]])

ev = np.linalg.eig(a) #固有値と固有ベクトルを同時に求める

print(ev[0]) #最初の要素は固有値

print()

print(ev[1]) #次の要素は固有ベクトル
```

演習

```python
import numpy as np

a = np. array([[-2, 4], [-1, 3]])

ev = np.linalg.eig(a)

print(ev[0]) #固有値

print()

print(ev[1]) #固有ベクトル
```
