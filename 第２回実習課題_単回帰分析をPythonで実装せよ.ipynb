統計学により、単回帰分析の最適な a と b は以下の式で求まることがわかっています。　
結論は簡単ですが、詳細な理論式の導出は少し難しいです。Web等に解説がたくさん載っています。　分散、共分散、単回帰分析で調べてみましょう。

数式：$$
\text{分散} = \frac{\sum_{i=1}^n (x_i-bar(x))^2}{n} 
$$

数式：$$
\text{共分散} = \frac{\sum_{i=1}^n (x_i-bar(x)) (y_i-bar(y))}{n} 
$$

数式：$$
\text{a} = \frac{\text{共分散}(x,y)}{\text{分散}(x)} 
$$

ここで、
*   $\bar{x}$, $\bar{y}$はそれぞれの平均値 (`np.mean`)
*   分散 (x) は `np.var`(`x, ddof=0`)
*   共分散 (x, y) は `np.cov`(`x, y, ddof=0`)[0, 1]
