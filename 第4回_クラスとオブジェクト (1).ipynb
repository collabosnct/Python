# 第4回：クラスとオブジェクト

より複雑なプログラムを整理して作るための「オブジェクト指向」という考え方の基礎です。
データと関数をひとまとめにして、「モノ」として扱います。

## 1. クラスとオブジェクト
* **クラス (`class`)**: どのようなデータと処理を持つかを定義した「設計図」です。
* **オブジェクト**: クラス（設計図）からメモリ上に作られた「実体」です。1つのクラスから複数のオブジェクトを作ることができます。

## 2. プロパティとメソッド
* **プロパティ**: そのオブジェクトが持っているデータ（変数）のことです。
* **メソッド**: そのオブジェクトが実行できる動作（クラス内に定義された関数）のことです。
* `self`: メソッドの最初の引数には必ず `self` を書きます。これは「オブジェクト自身」を指すキーワードです。

## 3. コンストラクタ (`__init__`)
オブジェクトが新しく作られるときに、一番最初に自動的に呼び出される特別なメソッドです。ここでプロパティの初期設定を行います。

```python
# --- 1. クラス（設計図）の定義 ---
class Car:
    # コンストラクタ: オブジェクトが作られる時の初期設定
    def __init__(self, model_name, color):
        self.model_name = model_name  # プロパティ（車種）
        self.color = color            # プロパティ（色）
        self.speed = 0                # 最初はスピード0

    # メソッド: 実行できる動作
    def accelerate(self, value):
        self.speed += value
        return f"{self.model_name}が加速しました。現在の時速は{self.speed}kmです。"

    def brake(self):
        self.speed = 0
        return f"{self.model_name}が停止しました。"

print("--- オブジェクトの作成と操作 ---")

# --- 2. オブジェクト（実体）の作成 ---
# Carクラスの設計図をもとに、2つの車（実体）を作る
my_car = Car("スポーツカー", "赤")
family_car = Car("ミニバン", "白")

# --- 3. プロパティの確認とメソッドの実行 ---
print(f"私の車は {my_car.color}色の {my_car.model_name} です。")
print(f"家族の車は {family_car.color}色の {family_car.model_name} です。\n")

# メソッドを呼び出して状態を変化させる
print(my_car.accelerate(40))
print(my_car.accelerate(20))
print(my_car.brake())

print("\n")

print(family_car.accelerate(30))
```
