転置の実装

```python
import numpy as np

a = np.array([[1, 2, 3], [4, 5, 6]]) #行列

print(a.T) #転置
```

転置と行列積の実装

```python
import numpy as np

a = np.array([[0, 1, 2], [1, 2, 3]]) #2x3の行列
b = np.array([[0, 1, 2], [1, 2, 3]]) #2x3の行列 このままでは行列積ができない

#print(np.dot(a, b)) #転置しないで行列積をとるとエラー
print(np.dot(a, b.T)) #転置により行列積が可能に
```

演習

```python
import numpy as np

a = np.array([[0, 1, 2], [1, 2, 3]])
b = np.array([[0, 1, 2], [2, 3, 4]])

#行列積
print(np.dot(a, b.T))
```
